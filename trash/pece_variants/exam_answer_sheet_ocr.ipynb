```python
# Local notebook wrapper around the implementation in src/.
from __future__ import annotations

import sys
from pathlib import Path

import cv2
import matplotlib.pyplot as plt
import numpy as np

def find_project_root(start: Path | None = None) -> Path:
    start = (start or Path.cwd()).resolve()
    candidates = [start, start / 'end-to-end', start.parent, start.parent / 'end-to-end']
    for candidate in candidates:
        if (candidate / 'src' / 'helper.py').exists() and (candidate / 'datasets').exists():
            return candidate
    for candidate in start.parents:
        if (candidate / 'src' / 'helper.py').exists() and (candidate / 'datasets').exists():
            return candidate
    raise FileNotFoundError('Could not locate the project root with src/ and datasets/.')

project_root = find_project_root()
datasets_root = project_root / 'datasets'
src_root = project_root / 'src'
for path in (project_root, src_root):
    if str(path) not in sys.path:
        sys.path.insert(0, str(path))

from src.helper import load_image, show_grid, show_image
from data.preprocessing import preprocess_image
from data.augmentation import build_augmentation_pipeline, generate_augmented_variants
from src.process_all import run_for_file

supported_suffixes = {'.png', '.jpg', '.jpeg', '.webp', '.bmp', '.tif', '.tiff', '.pdf', '.heic', '.heif'}
dataset_files = sorted(
    path for path in datasets_root.rglob('*')
    if path.is_file() and path.suffix.lower() in supported_suffixes
)
if not dataset_files:
    raise FileNotFoundError(f'No supported images found under: {datasets_root}')

selected_image_path = dataset_files[0]
print(f'Project root: {project_root}')
print(f'Datasets root: {datasets_root}')
print(f'Found {len(dataset_files)} supported files. Using: {selected_image_path}')

plt.rcParams['figure.figsize'] = (14, 10)
plt.rcParams['image.cmap'] = 'gray'
plt.rcParams['axes.titlesize'] = 11
plt.rcParams['axes.titleweight'] = 'bold'
```
